```python
import requests
import json
import quandl
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn import linear_model
from fbprophet import Prophet
quandl.ApiConfig.api_key = "psHSh62TY8dQsy4VXvAc"


data = quandl.get("EIA/PET_RWTC_W")


oil_df = pd.DataFrame(data).reset_index(drop=False)
oil_df
```

```python
oil_df.loc[:, "Date"] = oil_df["Date"].apply(lambda x: x + dt.timedelta(3))
```

```python
oil_df
```

```python
oil_df = oil_df.rename(columns={'Date':'ds','Value':'y'})
oil_df.head()
```

```python
m = Prophet(interval_width=.95)
model = m.fit(oil_df)
```

```python
future = m.make_future_dataframe(periods=26, freq='W')
forecast = m.predict(future)
forecast.tail()
```

```python
plot1 = m.plot(forecast)
```
